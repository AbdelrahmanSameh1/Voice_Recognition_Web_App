```python

from scipy.io.wavfile import read

directoryOpen = './RecordsData/open the door/'
directoryCloseThedoor = './RecordsData/close the door/'
directoryOwindow = './RecordsData/open the window/'
directoryCwindow = './RecordsData/close the window/'

openTheDoor = extractFromFile(directoryOpen)
closeTheDoor = extractFromFile(directoryCloseThedoor)
openTheWindow = extractFromFile(directoryOwindow)
closeTheWindow = extractFromFile(directoryCwindow)
```

```python
openthedoor_gmm = generateModel('openthedoor_gmm',openTheDoor, 'openTheDoor')
closethedoor_gmm = generateModel('closethedoor_gmm',closeTheDoor, 'closeTheDoor') 
openthewindow_gmm = generateModel('openthewindow_gmm',openTheWindow, 'openTheWindow')
closethewindow_gmm = generateModel('closethewindow_gmm',closeTheWindow, 'closeTheWindow')
```

```python
import pickle
import joblib
lstFolders = ['openTheDoor', 'closeTheDoor', 'openTheWindow', 'closeTheWindow']

models = [joblib.load(f + '.gmm') for f in lstFolders]
```

```python

import glob
from sklearn.metrics import f1_score

y = []
y_true = []
lstFolders = ['open the door' , 'close the door', 'open the window', 'close the window']

for folder in lstFolders:
    # try:
    lstFiles = glob.glob('RecordsData/{}/*'.format(folder))
    for f in lstFiles:
        
        sr,audio = read(f)
        vstack = extract_features(f)
        output = []
        for model in models:
            scores = np.array(model.score(vstack))
            output.append(scores.sum())

        winner = np.argmax(output)
        y_true.append(winner)
        y.append(lstFolders.index(folder))
```

```python
f1_score(y, y_true, average='micro')
```
